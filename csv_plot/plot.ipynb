```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
p = Path('.')

for kappa in [1, 2, 4, 8]:
	df = {}
	for csv_file in p.glob(f"k={kappa}*.csv"):
		_, method = csv_file.stem.split('-')
		csv = pd.read_csv(csv_file, usecols=['Value'])
		df[method] = csv.values.squeeze()
		# break

	df = pd.DataFrame(df)
	ordered_columns = ['Data', 'Res', 'Jac', 'Desc']
	styles = ['bs-','ro-','y^-']
	df = df[ordered_columns]
	df.rename(columns={'Desc':'$\mathcal{L}_{\mathcal{D}_3}$', 'Res':'$\mathcal{L}_{\\text{res}}$', 'Data':'$\mathcal{L}_{\\text{data}}$', 'Jac':'$\mathcal{L}_{\mathcal{J}_3}$'}, inplace=True)
	ax = df.plot(
		figsize=(14, 6),
		logy=True,
		fontsize=18,
		grid=True,
		linewidth=3,
		# style=styles
	)
	ax.set_title(f'Validation Errors of $\lambda = {kappa}$', fontsize=25)
	ax.set_xlabel('Epochs', fontsize=	20)
	ax.set_ylabel('$L_2$', 	fontsize=	20)
	
	plt.xticks(fontsize=12)
	plt.yticks(fontsize=12)
	ax.legend(df.columns, 	fontsize=	20, title_fontsize=20, ncol=2)

	# Show the plot
	plt.tight_layout()

	plt.savefig(f'./Val-{kappa}.png')
	plt.close()
```
