```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
p = Path('.')

for kappa in [1, 2, 4, 8]:
	df = {}
	for csv_file in p.glob(f"k={kappa}*.csv"):
		_, method = csv_file.stem.split('-')
		csv = pd.read_csv(csv_file, usecols=['Value'])
		df[method] = csv.values.squeeze()
		# break

	df = pd.DataFrame(df)
	ordered_columns = ['Data', 'Res', 'Jac', 'Desc']
	df = df[ordered_columns]
	df.rename(columns={'Desc':'$\mathcal{L}_{\mathcal{D}^3}$', 'Res':'Residual', 'Data':'Data', 'Jac':'$\mathcal{L}_{\mathcal{J}^3}$'}, inplace=True)
	ax = df.plot(
		figsize=(14, 6),
		logy=True,
		fontsize=17,
		grid=True,
		linewidth=2,
		
	)
	ax.set_title(f'Validation Errors of $\kappa = {kappa}$', fontsize=19)
	ax.set_xlabel('Epochs', fontsize=19)
	ax.set_ylabel('$L_2$', fontsize=19)
	ax.legend(df.columns, fontsize=18, title_fontsize=14, ncol=2, )

	# Show the plot
	plt.tight_layout()

	plt.savefig(f'./Val-{kappa}.png')
	plt.close()
```
